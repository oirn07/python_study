### Beautiful Soup 모듈
- 파이썬의 웹 페이지 분석 모듈

```python
# Beautiful Soup 모듈로 날씨 가져오기
from urllib import request
from bs4 import BeautifulSoup

target = request.urlopen("https://www.korea.kr/rss/dept_kasa.xml")

soup = BeautifulSoup(target, "html.parser")

for item in soup.select("item"):
    title = item.select_one("title").text
    guid  = item.select_one("guid").text
    desc  = item.select_one("description").text
    date  = item.select_one("pubDate").text
    print(title, guid, date)
```
